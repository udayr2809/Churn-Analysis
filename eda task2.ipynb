### Separating Numeric and Categorical column

```python
pricenumcol=[]
pricecatcol=[]
for col in price.columns:
    if price[col].dtype=='o' or price[col].dtype=='object':
        pricecatcol.append(col)
    else:
        pricenumcol.append(col)
```

```python
pricecatcol
```

```python
pricenumcol
```

```python
clientnumcol=[]
clientcatcol=[]
for col in client.columns:
    if client[col].dtype=='o' or client[col].dtype=='object':
        clientcatcol.append(col)
    else:
        clientnumcol.append(col)
```

```python
clientcatcol
```

```python
clientnumcol
```

### Visualizing Distributions of columns 

#### Client Dataset

```python
for i in clientnumcol:
    plot = sns.displot(client[i], kde=True, height=4, aspect=2)
    plot.set(title="Distribution of {} in data".format(i))
    plt.xlabel(i)
    plt.ylabel('Frequency')
```

#### Price Dataset

```python
for i in pricenumcol:
    plot = sns.displot(price[i], kde=True, height=4, aspect=2)
    plot.set(title="Distribution of {} in data".format(i))
    plt.xlabel(i)
    plt.ylabel('Frequency')
```

#### Categorical Column Distribution

```python
# Distribution of channel sales
```

```python
data = client['channel_sales'].value_counts()
labels=client['channel_sales'].unique()
```

```python
plt.figure(figsize=(4, 4))
plt.pie(data, labels=labels)
plt.title('Distribution of Channel Sales')
```

```python
# Distribution gas client
```

```python
data = client['has_gas'].value_counts()
labels=client['has_gas'].unique()
plt.figure(figsize=(4, 4))
plt.pie(data, labels=labels)
plt.title('Distribution of Gas Client')
```

```python
#Distribution of client churned over the next 3 months
```

```python
data = client['churn'].value_counts()
labels=client['churn'].unique()
plt.figure(figsize=(4, 4))
plt.pie(data, labels=labels)
plt.title('Distribution of client churned over the next 3 months')
```
